## Cargar y visualizar datos

## Filtrar computadoras usadas

## Computadora más cara

```python
df = pd.read_csv("computadoras.csv")
computadora_mas_cara = df.loc[df['Precio'].idxmax()]
print("La computadora más cara es:")
print(computadora_mas_cara)
```

## Agregar columna calculada

```python
df = pd.read_csv("computadoras.csv")
df['Precio por GB'] = df['Precio'] / df['Almacenamiento (GB)']
print("Dataset con la nueva columna 'Precio por GB':")
print(df.head())
```

# Gráficos

## Gráfico de barras: Promedio de precios por marca

```python
df = pd.read_csv("computadoras.csv")
precio_promedio = df.groupby('Marca')['Precio'].mean()
plt.figure(figsize=(10, 6))
precio_promedio.plot(kind='bar', color='skyblue')
plt.title('Precio Promedio de las Computadoras por Marca')
plt.xlabel('Marca')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Gráfico de líneas: Tendencia de precios promedio por año

```python
df = pd.read_csv("computadoras.csv")
precio_promedio_anual = df.groupby('Año')['Precio'].mean()
plt.figure(figsize=(10, 6))
precio_promedio_anual.plot(kind='line', marker='o', color='b')
plt.title('Cambio del Precio Promedio de las Computadoras a lo Largo de los Años')
plt.xlabel('Año')
plt.ylabel('Precio Promedio')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Gráfico circular: Proporción de estados de computadoras

```python
df = pd.read_csv("computadoras.csv")
estado_computadoras = df['Estado'].value_counts()
plt.figure(figsize=(8, 8))
estado_computadoras.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Proporción de Computadoras por Estado')
plt.ylabel('')
plt.show()
```

## Diagrama de dispersión: Relación entre precio y almacenamiento

```python
df = pd.read_csv("computadoras.csv")
plt.figure(figsize=(10, 6))
plt.scatter(df['Almacenamiento (GB)'], df['Precio'], color='purple', alpha=0.5)
plt.title('Relación entre el Precio y la Capacidad de Almacenamiento')
plt.xlabel('Capacidad de Almacenamiento (GB)')
plt.ylabel('Precio')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Histograma: Distribución de precios de computadoras usadas

```python
df = pd.read_csv("computadoras.csv")
computadoras_usadas = df[df['Estado'] == 'Usado']
plt.figure(figsize=(10, 6))
plt.hist(computadoras_usadas['Precio'], bins=20, color='orange', edgecolor='black')
plt.title('Distribución de Precios de Computadoras Usadas')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.tight_layout()
plt.show()
```
